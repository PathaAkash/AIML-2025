```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
Cancer = load_breast_cancer()
Cancer.keys()
```

```python
Cancer['feature_names']
```

```python
df = pd.DataFrame(Cancer['data'],columns=Cancer['feature_names'])
df.head()

```

```python
df["target"] = Cancer["target"]
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
X= df.drop(labels = ['target'], axis = 1)
y = df['target']
X.shape
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state=100)
print("Training Data ",X_train.shape)
print("Testing Data",X_test.shape)

```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
svm.fit(X_train, y_train)
```

```python
y_predicted_train = svm.predict(X_train)
y_predicted_test = svm .predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print("Training Accuracy = ", accuracy_score(y_train, y_predicted_train))
print("Testing Accuracy = ", accuracy_score(y_test, y_predicted_test))
```

```python
print(classification_report(y_train, y_predicted_train))
```

```python
cm = confusion_matrix(y_train,y_predicted_train)
cm
```

```python
sns.heatmap(cm, annot = True)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1, 10 , 100, 1000],
              'gamma':[1,0.1,0.01,0.001,0.0001],
              'kernel' : ['rbf']}
```

```python
grid_model = GridSearchCV(SVC(),param_grid,refit=True,verbose=1)

```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_params_
```

```python
grid_predictions = grid_model.predict(X_test)
```

```python
print("Testing Accuracy = ", accuracy_score(y_test,grid_predictions))

```

```python
cm = confusion_matrix(y_test, grid_predictions)
cm

```

```python
sns.heatmap(cm, annot = True  ,cmap = 'viridis_r')
```
