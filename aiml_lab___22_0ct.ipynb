```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer.keys()


```

```python
cancer['feature_names']
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
df['target'] = cancer['target']
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
x = df.drop(labels = ['target'],axis = 1)
y = df['target']
x.shape
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=100)
print("Training data",x_train.shape)
print("testing data ",x_test.shape)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
y_predicted_train = svm.predict(x_train)
y_predicted_test = svm.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print("Training Accuracy",accuracy_score(y_train,y_predicted_train))
print("Testing Accuracy",accuracy_score(y_test,y_predicted_test))
```

```python
print(classification_report(y_test, y_predicted_test))
```

```python
cm=confusion_matrix(y_train,y_predicted_train)
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001],'kernel':['rbf']}
```

```python
grid_model=GridSearchCV(SVC(),param_grid,refit=True,verbose=1)
```

```python
grid_model.fit(x_train,y_train)
```

```python
grid_model.best_params_
```

```python
grid_predictions=grid_model.predict(x_test)
```

```python
print("testing Accuracy = ", accuracy_score(y_test,grid_predictions))
```

```python
print("testing Accuracy = ", accuracy_score(y_test,grid_predictions))
```
